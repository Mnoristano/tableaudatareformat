```python
import pandas as pd
```

```python
may18 = pd.read_csv('raw_data/201805-citibike-tripdata.csv') 
jun18 = pd.read_csv('raw_data/201806-citibike-tripdata.csv') 


nov18 = pd.read_csv('raw_data/201811-citibike-tripdata.csv') 
dec18 = pd.read_csv('raw_data/201812-citibike-tripdata.csv') 
```

```python
cold = [nov18,dec18]
warm = [may18, jun18]

cold18 = pd.concat(cold, axis = 0, ignore_index = True)
warm18 = pd.concat(warm, axis = 0, ignore_index = True)
```

```python
cold18['weather'] = 'cold'
warm18['weather'] = 'warm'

weatherlist = [cold18,warm18]
```

```python
alldata = pd.concat(weatherlist, axis = 0, ignore_index = True)
```

```python
alldata.to_csv('alldata2018.csv')
```
